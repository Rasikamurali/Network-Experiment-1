```python
import networkx as nx 
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
G = nx.barabasi_albert_graph(100,2)
```

```python
nx.draw_spring(G)
```

```python
plt.hist([v for k,v in nx.degree(G)])
```

The above shows that we have 75+ nodes that have less than 5 stubs for connection, compared to one node that has almost 25-ish stubs for connection. 

```python
plt.hist(nx.centrality.betweenness_centrality(G).values())
```

Centrality shows us how much influence a node has. As we can see the multiple nodes that had very few connections, have low influence/ centrality. The node with the most amount of connections as the most influence. 

```python
nx.diameter(G)
```

The minimum number of hops needed to go from one end to another

```python
nx.cluster.average_clustering(G)
```

The probability that if A is connected to B and A is connected to C, then B is also connected to C.
